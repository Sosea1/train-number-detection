```python
# Функция рабитие базы в процентном соотношении
def percent_data(images_dir_path = images_dir_path, a=75, b=15, c=10):
  count_images = len(os.listdir(images_dir_path))
  percent_train = round(count_images/100*a)
  percent_test = round(count_images/100*b)
  percent_val = round(count_images/100*c)
  return percent_train, percent_test, percent_val
```

```python
images = os.listdir(images_dir_path) # Каталог с картинками полный
labels = os.listdir(labels_dir_path) # Каталог с масками
size = 1080
# script for build data catalog with resize images
iter = 0
percent_train, percent_test, percent_val = percent_data()
for label in labels:
  pref_lbl = label.split(sep='.')[0]

  for image in images:
    pref_img = image.split(sep='.')[0]

    if pref_lbl == pref_img and iter < percent_train:
      iter += 1
      print(iter)
      print(pref_lbl, pref_img)

      file_path_images = images_dir_path + pref_img +'.jpg'
      print(file_path_images)

      # resize image
      image_old_size = cv2.imread(file_path_images)
      image_old_size = cv2.cvtColor(image_old_size, cv2.COLOR_BGR2RGB)
      img_h,img_w, _ = image_old_size.shape
      new_size = (size, size)
      image_new_size = cv2.resize(image_old_size, new_size)
      img_h_resize, img_w_resize, _ = image_new_size.shape

      # save image
      %cd /content/drive/MyDrive/Hackaton/YOLO/data/train/images
      filename = pref_img  + '.jpg' # give name file
      cv2.imwrite(filename, image_new_size)

      # work in labels
      path_to_labels = labels_dir_path + pref_lbl + '.json'
      print(path_to_labels)

      with open(path_to_labels) as f:
        templates = json.load(f)

      coord =  templates[0]['annotations'][0]['coordinates']
      coord = list(coord.values())

      x, y, w, h = coord
      x_r = x*(img_w_resize/img_w)
      y_r = y*(img_h_resize/img_h)
      w_r = w*(img_w_resize/img_w)
      h_r = h*(img_h_resize/img_h)

      x_r = x_r / img_w_resize
      y_r = y_r / img_h_resize
      w_r = w_r / img_w_resize
      h_r = h_r / img_h_resize

      file_name_txt =  pref_lbl +'.txt'
      %cd /content/drive/MyDrive/Hackaton/YOLO/data/train/labels
      with open(file_name_txt, "w") as file:
        file.write(f'{0} {x_r} {y_r} {w_r} {h_r}')

    elif pref_lbl == pref_img and iter < percent_train + percent_test:
      iter += 1
      print(iter)
      print(pref_lbl, pref_img)

      file_path_images = images_dir_path + pref_img +'.jpg'
      print(file_path_images)

      # resize image
      image_old_size = cv2.imread(file_path_images)
      image_old_size = cv2.cvtColor(image_old_size, cv2.COLOR_BGR2RGB)
      img_h,img_w, _ = image_old_size.shape
      new_size = (size, size)
      image_new_size = cv2.resize(image_old_size, new_size)
      img_h_resize, img_w_resize, _ = image_new_size.shape

      # save image
      %cd /content/drive/MyDrive/Hackaton/YOLO/data/test/images
      filename = pref_img  + '.jpg' # give name file
      cv2.imwrite(filename, image_new_size)

      # work in labels
      path_to_labels = labels_dir_path + pref_lbl + '.json'
      print(path_to_labels)

      with open(path_to_labels) as f:
        templates = json.load(f)

      coord =  templates[0]['annotations'][0]['coordinates']
      coord = list(coord.values())

      x, y, w, h = coord
      x_r = x*(img_w_resize/img_w)
      y_r = y*(img_h_resize/img_h)
      w_r = w*(img_w_resize/img_w)
      h_r = h*(img_h_resize/img_h)

      x_r = x_r / img_w_resize
      y_r = y_r / img_h_resize
      w_r = w_r / img_w_resize
      h_r = h_r / img_h_resize

      file_name_txt =  pref_lbl +'.txt'
      %cd /content/drive/MyDrive/Hackaton/YOLO/data/test/labels
      with open(file_name_txt, "w") as file:
        file.write(f'{0} {x_r} {y_r} {w_r} {h_r}')

    elif pref_lbl == pref_img and iter < percent_train + percent_test + percent_val:
      iter += 1
      print(iter)
      print(pref_lbl, pref_img)

      file_path_images = images_dir_path + pref_img +'.jpg'
      print(file_path_images)

      # resize image
      image_old_size = cv2.imread(file_path_images)
      image_old_size = cv2.cvtColor(image_old_size, cv2.COLOR_BGR2RGB)
      img_h,img_w, _ = image_old_size.shape
      new_size = (size, size)
      image_new_size = cv2.resize(image_old_size, new_size)
      img_h_resize, img_w_resize, _ = image_new_size.shape

      # save image
      %cd /content/drive/MyDrive/Hackaton/YOLO/data/valid/images
      filename = pref_img  + '.jpg' # give name file
      cv2.imwrite(filename, image_new_size)

      # work in labels
      path_to_labels = labels_dir_path + pref_lbl + '.json'
      print(path_to_labels)

      with open(path_to_labels) as f:
        templates = json.load(f)

      coord =  templates[0]['annotations'][0]['coordinates']
      coord = list(coord.values())

      x, y, w, h = coord
      x_r = x*(img_w_resize/img_w)
      y_r = y*(img_h_resize/img_h)
      w_r = w*(img_w_resize/img_w)
      h_r = h*(img_h_resize/img_h)

      x_r = x_r / img_w_resize
      y_r = y_r / img_h_resize
      w_r = w_r / img_w_resize
      h_r = h_r / img_h_resize

      file_name_txt =  pref_lbl +'.txt'
      %cd /content/drive/MyDrive/Hackaton/YOLO/data/valid/labels
      with open(file_name_txt, "w") as file:
        file.write(f'{0} {x_r} {y_r} {w_r} {h_r}')

```
